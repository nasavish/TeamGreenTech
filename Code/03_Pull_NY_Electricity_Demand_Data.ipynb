This was an attempt to pull in NY eletricity demand and solar farm generation data to use with the NY solar rooftop analysis. A few notes.

* NY didn't povide the solar farm generation data
* We are able to successfully pull and load the NY electric demand data, which is on hourly basis
* There is a total of 25.7K observations
* This amounts to bit less than 3 years of data, far shorter than the 18 years of data in the solar installation data
* We could still compare how rooftop solar production compares to overall electricity demand. Even if it short time frame, it may be instructive and make sure of the API. 
* We do need additional time and expertise to deal with the date time.stuff.

```python
import json
import pandas as pd
import apikeys
import requests
from pprint import pprint
```

```python
#Read in Net Generation Data
url = f'http://api.eia.gov/series/?api_key={apikeys.eia_key}&series_id=EBA.NYIS-ALL.D.H'
response = requests.get(url).json()
pprint(response)
```

```python
#Generate dataset
data = response['series'][0]['data']

#Go through each row and include Date and Demand
datetime = []
mwh = []
for row in data:
    datetime.append(row[0])
    mwh.append(row[1])


#Convert it into a DataFrame
data_dict ={ 
    "datetime_str": datetime,
    "mwh": mwh
    }

data = pd.DataFrame(data_dict)
data.head()
```

```python
#Extract Year, month and date (Note: I'm not sure these are in Eastern Time)
data['year'] = data.datetime_str.str[:4]
data['month']= data.datetime_str.str[4:6]
data['day']= data.datetime_str.str[6:8]
data['hour']=data.datetime_str.str[9:11]
 
for var in ['year', 'month', 'day', 'hour']:
    data[var] = pd.to_numeric(data[var])
    
```
